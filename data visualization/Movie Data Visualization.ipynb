### Data Visualization Challenge
#### Given a movie data set,your task is to make visualization that :
#### Plots the length of movies title names on x-axis
#### Frequency Counts on y axis ie number of movies having 'x' character in thier title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read a csv file
df = pd.read_csv('movie_metadata.csv')
```

```python
df.head(n=10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
print(titles[:5])
```

```python
freq_titles = {}
for title in titles:
    length = len(title)
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
x = np.array(list(freq_titles.keys()))
y = np.array(list(freq_titles.values()))
print(x,y)
```

```python
plt.style.use('seaborn-v0_8')
plt.scatter(x,y)
plt.xlabel('Length of Movie Title')
# plt.ylabel('No of Times a movies has this length')
plt.ylabel('Frequency Count')
plt.title('Movie Data Visualization Problem')
plt.show()
```
