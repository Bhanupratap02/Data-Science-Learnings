### Data Visualization using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use('seaborn-v0_8')
```

```python
#line plot
x = np.arange(10)
y1 = x*2
y2 = 2*x+3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color='red',label='Apple',marker='o')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruits over time')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(4,4))
plt.scatter(x,y1,color='red',label='Apple',marker='o')
plt.scatter(x,y2,color='green',label='Kiwi',linestyle='dashed',marker='*')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Scatter Plot :Prices of fruits over time')
plt.legend()
plt.show()
```

```python
x=np.array([0,1,2])*2
plt.bar(x-0.25,[10,20,15],width=0.5,label='Current Year',tick_label=['Gold','Silver','Platinum'])
plt.bar(x+0.25,[20,13,12],width=0.5,label='Next year')
plt.title('Metal Price Comparision')
plt.xlabel('Metal')
plt.ylabel('Price')
plt.ylim(0,40)
plt.legend()
plt.show()
```

```python
subjects = "Maths","Physics","Chem","English"
weights = [20,10,15,5]
plt.pie(weights,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
xns = np.random.randn(100)
sigma = 5
u = 60
x = np.round(xns*sigma + u)
x2 = np.round(xns*sigma + 40)
print(x,x2)
```

```python
plt.hist(x,alpha=0.9,label='Maths')
plt.hist(x2,alpha=0.8,label='Physics')
plt.xlabel('Marks Range')
plt.ylabel('Frequency Count of Students')
plt.title('Histogram')
plt.legend()
plt.show()
```
